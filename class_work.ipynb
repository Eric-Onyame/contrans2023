```python
states = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
root = 'http://www.opensecrets.org/api'
df_total = pd.DataFrame()
for i in states:
    endpoint  = f"/?method=getLegislators&id={i}"
    print(root+endpoint)
    parameters = {'output':'json',
                  'apikey': opensecretsapi
                 }
    r = requests.get(root+endpoint,
                params = parameters)
    stateLegislators = json.loads(r.text)
    data = stateLegislators["response"]["legislator"]
    new_data = []
    for j in data:
        new_data.append(j["@attributes"])
    df_temp  = pd.DataFrame.from_dict(new_data)

 

    df_total = pd.concat([df_total, df_temp],ignore_index = True)
    
# root = 'http://www.opensecrets.org/api'
# for i in states:
#     #state=i
#     endpoint=f'/?method=getLegislators&id={i}'
#     print(root+endpoint)
#     parameters = {'output':'json',
#              'apikey': opensecretsapi}
#     r= requests.get (root+endpoint,
#                     params=parameters)
#     print(r)
#     statelegislators=json.loads(r.text)
#     data=statelegislators["response"]["legislator"]
#     new_data=[]
#     for j in data:
#         new_data.append(j["@attributes"])
#     df_temp =pd.DataFrame.from_dict(new_data)

#     df_total = pd.concat(df_total,)
#     #print(r)


#https://www.opensecrets.org/api/?method=candContrib&cid=N00007360&cycle=2022&apikey=__apikey__
```
